```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
def recommend(product):
    product_index = df[df["Product_Sub_Category"] == product].index[0]
    distances = similarity[product_index]
    product_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:100]
    
    printed_products = set()  # Set to store printed product names
    printed_count = 0  # Counter for printed products
    
    for i in product_list:
        product_name = df.iloc[i[0]].Product_Name
        if product_name not in printed_products:  # Check if product has not been printed before
            print(product_name)
            printed_products.add(product_name)  # Add printed product name to set
            printed_count += 1  # Increment printed count
            if printed_count >= 5:  # Exit loop if 5 products have been printed
                break
```
